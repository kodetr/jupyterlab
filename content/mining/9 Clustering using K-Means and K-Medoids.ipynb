Algoritme ini mengikuti cara yang mudah atau sederhana untuk mengklasifikasikan kumpulan data yang diberikan melalui sejumlah cluster, apriori tetap. ## <b>6 <span style='color:#e61227'>|</span> Perhitungan K-means</b>

# Scalling

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

selected_columns = ['jumlah kejahatan (2021)', 'persentase korban (2021)']
X = df[selected_columns]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled.head()
```

# Klasterisasi/Clustering

### Nilai K Awal

```python
kmean= KMeans(n_clusters=3)
y_cluster = kmean.fit_predict(X_scaled )
y_cluster
```

```python
df['Cluster (2021)']=y_cluster
df
```

### Visualisasi Hasil Clustering

```python
# Visualisasi hasil clustering dengan K-Means
plt.figure(figsize=(10, 6))
colors = ['blue', 'green', 'orange']  # Tentukan warna untuk setiap cluster
for i in range(3):
  plt.scatter(X_scaled['jumlah kejahatan (2021)'][df['Cluster (2021)'] == i],
              X_scaled['persentase korban (2021)'][df['Cluster (2021)'] == i],
              c=colors[i], label=f'Cluster {i}')

plt.scatter(kmean.cluster_centers_[:, 0], kmean.cluster_centers_[:, 1], marker='*', s=200, c='red', label='Centroids')
plt.xlabel('Jumlah Kejahatan')
plt.ylabel('Persentase Korban')
plt.title('Hasil Clustering K-Means')
plt.legend()
plt.show()
```

### Centroid Awal

```python
centroids = kmean.cluster_centers_
print(f"\nNilai Centroid untuk Nilai K awal:")
for i, centroid in enumerate(centroids):
    print(f"Centroid-{i+1}: {centroid}")
```

### mencari niali k terbaik menggunakan ebow method

```python
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Method
plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
from sklearn_extra.cluster import KMedoids

# Inisialisasi K-Medoids
kmedoids = KMedoids(n_clusters=3, random_state=0)

# Melakukan clustering
y_kmedoids = kmedoids.fit_predict(X_scaled)
df['Cluster_KMedoids (2021)'] = y_kmedoids
df
```

```python
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_score

silhouette_avg = []
for k in range(2, 10):
    kmedoids = KMedoids(n_clusters=k)
    cluster_labels = kmedoids.fit_predict(X_scaled)
    silhouette_avg.append(silhouette_score(X_scaled, cluster_labels))

# Plot Silhouette Analysis
plt.plot(range(2, 10), silhouette_avg, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal K')
plt.show()
```

Inertia dan Silhouette Cluster adalah dua metrik yang digunakan untuk mengevaluasi kualitas clustering, tetapi mereka memiliki pendekatan yang berbeda.



> Inertia:



* Definisi: Jumlah total jarak kuadrat antara setiap titik data dan centroid clusternya.
* Algoritma: Digunakan terutama dalam algoritma K-Means.
* Interpretasi:
Nilai inertia yang lebih rendah umumnya menunjukkan cluster yang lebih baik, karena titik data lebih dekat ke pusat cluster mereka.
Namun, inertia cenderung menurun seiring bertambahnya jumlah cluster, sehingga tidak selalu menjadi indikator terbaik untuk menentukan jumlah cluster yang optimal.
* Keterbatasan:
Sensitif terhadap skala data dan bentuk cluster.
Tidak cocok untuk cluster dengan bentuk non-globular.



> Silhouette Cluster:



* Definisi: Mengukur seberapa mirip suatu titik data dengan clusternya sendiri dibandingkan dengan cluster lain.
* Algoritma: Dapat digunakan dengan berbagai algoritma clustering, termasuk K-Means dan K-Medoids.
* Interpretasi:
Nilai silhouette berkisar antara -1 hingga 1.
Nilai mendekati 1 menunjukkan bahwa titik data ditempatkan dengan baik di clusternya.
Nilai mendekati 0 menunjukkan bahwa titik data berada di dekat batas antara dua cluster.
Nilai mendekati -1 menunjukkan bahwa titik data mungkin telah ditempatkan di cluster yang salah.
* Keuntungan:
Lebih robust terhadap bentuk cluster yang berbeda.
Dapat memberikan informasi yang lebih detail tentang kualitas masing-masing cluster.

> Analogi:

Bayangkan Anda mengelompokkan siswa berdasarkan nilai ujian mereka.

Inertia: Akan mengukur total jarak kuadrat antara nilai setiap siswa dan nilai rata-rata kelompoknya.
Silhouette: Akan mengukur seberapa dekat nilai seorang siswa dengan nilai siswa lain di kelompoknya sendiri dibandingkan dengan nilai siswa di kelompok lain.
Kesimpulan:

Inertia adalah ukuran kekompakan cluster, sedangkan silhouette cluster adalah ukuran kebaikan penempatan titik data dalam cluster.
Silhouette cluster umumnya dianggap sebagai metrik yang lebih komprehensif untuk mengevaluasi kualitas clustering, terutama ketika berhadapan dengan cluster non-globular atau data dengan skala yang berbeda.
