> for loop
Digunakan untuk iterasi atas suatu koleksi (seperti list, tuple, string, dll.) atau rentang angka menggunakan range().

Dapat digunakan untuk melakukan iterasi berdasarkan indeks atau elemen.

```python
#Looping pada list:
daftar_hobi = ["Bersepeda", "Membaca", "Memasak"]
for hobi in daftar_hobi:
    print(hobi)
```

```python
# Looping pada range (bilangan):
for i in range(5):  # Loop dari 0 sampai 4
    print(i)
```



> while loop
Digunakan untuk mengulang suatu blok kode selama kondisi tertentu bernilai True.

Lebih cocok untuk situasi di mana jumlah iterasi tidak diketahui sebelumnya, dan kondisi perlu dicek secara dinamis selama proses berlangsung.
Namun, meskipun Python hanya memiliki dua loop utama (for dan while), ada fitur dan konsep tambahan yang terkait dengan pengulangan dan iterasi, seperti:

Konstruksi Terkait dengan Loop:

* break: Keluar dari loop sebelum waktunya, menghentikan eksekusi loop.
* continue: Melewatkan iterasi saat ini dan melanjutkan ke iterasi berikutnya.
* else pada loop: Blok else dijalankan ketika loop selesai secara normal (tanpa break).

```python
# For Loop dengan break dan continue:
for i in range(5):
    if i == 3:
        break  # Keluar dari loop jika i sama dengan 3
    print(i)

print(50*'=')

for i in range(5):
    if i == 2:
        continue  # Lewati iterasi jika i sama dengan 2
    print(i)
```

```python
# While Loop dengan break dan continue:
angka = 0

while angka < 5:
    print(angka)
    if angka == 3:
        break  # Keluar dari loop jika angka sama dengan 3
    angka += 1

print(50*'=')

angka = 0
while angka < 5:
    if angka == 2:
        angka += 1
        continue  # Lewati iterasi jika angka sama dengan 2
    print(angka)
    angka += 1
```

# Basic Numpy
NumPy (Numerical Python)
* NumPy adalah library Python yang digunakan untuk komputasi ilmiah dan operasi numerik. Fungsi utama dari NumPy adalah menyediakan array multidimensi (ndarray) dan alat-alat untuk memanipulasinya dengan cara yang sangat efisien.

```python
import numpy as np

# Membuat array NumPy
arr = np.array([1, 2, 3, 4, 5])

# Operasi pada array
print("Array:", arr)
print("Jumlah elemen:", arr.sum())
print("Rata-rata elemen:", arr.mean())
print("Elemen terbesar:", arr.max())
print("Elemen terkecil:", arr.min())
```

```python
# Indexing dan slicing
print("Elemen ke-2:", arr[1])
print("Elemen dari indeks 1 hingga 3:", arr[1:4])
```

```python
# Operasi matriks
matrix = np.array([[1, 2], [3, 4]])
print("Matriks:\n", matrix)
print("Transpose matriks:\n", matrix.transpose())
print("Determinan matriks:", np.linalg.det(matrix))
```

# Basic Pandas
Pandas adalah library yang digunakan untuk manipulasi dan analisis data. Pandas menyediakan dua struktur data utama: DataFrame (data tabular) dan Series (data satu dimensi), yang sangat kuat dan mudah digunakan untuk bekerja dengan data berbentuk tabel (misalnya data CSV, Excel, SQL, dll.).

```python
import pandas as pd

# Membuat DataFrame dari dictionary
data = {'Nama': ['Budi', 'Ani', 'Citra'],
        'Usia': [25, 30, 28],
        'Kota': ['Jakarta', 'Bandung', 'Surabaya']}
df = pd.DataFrame(data)
print("DataFrame:\n", df)
df
```

```python
# Mengakses kolom
print("\nKolom Nama:\n", df['Nama'])
df['Nama']
```

```python
# Mengakses baris
print("\nBaris ke-2:\n", df.loc[1])
df.loc[1]
```

```python
# Filtering data
print("\nData dengan usia di atas 28:\n", df[df['Usia'] > 28])
df[df['Usia'] > 28]
```

```python
# Menambahkan kolom baru
df['Pekerjaan'] = ['Dokter', 'Guru', 'Insinyur']
print("\nDataFrame dengan kolom Pekerjaan:\n", df)
df
```

```python
# Menghapus kolom
df = df.drop('Kota', axis=1)
print("\nDataFrame setelah menghapus kolom Kota:\n", df)
df
```

```python
# Mengurutkan data
df_sorted = df.sort_values('Usia')
print("\nDataFrame setelah diurutkan berdasarkan usia:\n", df_sorted)
df_sorted
```

```python
# Agregasi data
print("\nRata-rata usia:\n", df['Usia'].mean())
```

```python
# Groupby
data = {'Nama': ['Budi', 'Ani', 'Citra', 'Deni', 'Eka'],
        'Usia': [25, 30, 28, 25, 30],
        'Kota': ['Jakarta', 'Bandung', 'Surabaya', 'Jakarta', 'Bandung']}
df = pd.DataFrame(data)
df
```

```python
grouped = df.groupby('Kota')['Usia'].mean()
grouped
```

# Basic Matplotlib
Matplotlib adalah library yang digunakan untuk visualisasi data. Matplotlib sangat fleksibel dan memungkinkan pembuatan berbagai jenis grafik dan plot, seperti garis, batang, scatter plot, histogram, dan sebagainya.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [25, 32, 30, 35, 27]
values2 = [18, 28, 25, 30, 22]

# Membuat subplot
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot untuk values1
bar_width = 0.35
bar_positions = np.arange(len(categories))
ax.bar(bar_positions, values1, width=bar_width, label='Data 1', color='skyblue')

# Bar plot untuk values2
ax.bar(bar_positions + bar_width, values2, width=bar_width, label='Data 2', color='lightcoral')

# Menambahkan label sumbu
ax.set_xlabel('Kategori')
ax.set_ylabel('Nilai')
ax.set_title('Perbandingan Data 1 dan Data 2')

# Menambahkan label kategori pada sumbu x
ax.set_xticks(bar_positions + bar_width / 2)
ax.set_xticklabels(categories)

# Menambahkan legenda
ax.legend()

# Menambahkan grid
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Menambahkan anotasi pada setiap bar
for i, v in enumerate(values1):
    ax.text(bar_positions[i], v + 0.5, str(v), ha='center', va='bottom')
for i, v in enumerate(values2):
    ax.text(bar_positions[i] + bar_width, v + 0.5, str(v), ha='center', va='bottom')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

# Scikit-learn (sklearn)
Scikit-learn adalah library yang sangat populer untuk machine learning dan data mining di Python. Library ini menyediakan alat-alat yang digunakan untuk membangun dan menguji model machine learning.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load dataset iris
iris = load_iris()
X = iris.data  # Fitur
y = iris.target  # Target (label)

# Bagi dataset menjadi data training dan data testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Buat model KNN
knn = KNeighborsClassifier(n_neighbors=3)

# Training model dengan data training
knn.fit(X_train, y_train)

# Prediksi label untuk data testing
y_pred = knn.predict(X_test)

# Hitung akurasi model
accuracy = accuracy_score(y_test, y_pred)
print("Akurasi model:", accuracy)
```
