```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_path = "daun.png"

print(image_path)
```

```python
threshold=127
```

```python
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
_, binary_image = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image, cmap="gray")

plt.subplot(1, 2, 2)
plt.title("Segmentasi")
plt.imshow(binary_image, cmap="gray")

plt.tight_layout()
plt.show()
```

```python
# Import library yang dibutuhkan
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Membaca citra dari file
image = cv2.imread('daun.png', 0)  # Membaca dalam format grayscale

# Menerapkan Thresholding
ret, thresh1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(image, 1, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(image, 1, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(image, 1, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(image, 1, 255, cv2.THRESH_TOZERO_INV)

# Menampilkan hasil segmentasi
titles = ['Original Image', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [image, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```
