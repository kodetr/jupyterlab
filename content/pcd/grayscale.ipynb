# Convert Grayscale

```python
for y in range(img.shape[0]) : # Y 921
    for x in range(img.shape[1]) : # X 614

        blue = img[y,x,0]
        green = img[y,x,1]
        red = img[y,x,2]

        grayscale = (0.5 * red) + (0.3 * green) + (0.2 * blue)

        img[y,x,0] = grayscale # blue
        img[y,x,1] = grayscale # green
        img[y,x,2] = grayscale # red

plt.imshow(img)
plt.show()
```
